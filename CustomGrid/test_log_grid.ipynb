```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
sns.set_theme()
sns.set_style("ticks")
sns.set_context("paper")
```

```python
def is_log(x, y, log_vars=None, **kwargs):
    # Map log scales
    # pass
    # Get current axis
    ax = plt.gca()

    if x.name in log_vars:
        ax.set_xscale('log')
    if y.name in log_vars:
        ax.set_yscale('log')
    ax.autoscale()
```

```python
data_path = '../Data/Spitzer_ALMA_sample.csv'
spitzer = pd.read_csv(data_path, sep=',', skipinitialspace=True, na_values=['#NAME?'])
```

```python
variables = ['MSTAR', 'LSTAR', 'LOGLACC', 'LOGMACC', 'TEFF', 'N1331', 'DIST', 'INCL']
plot_log = ['MSTAR', 'LSTAR', 'DIST', 'FLC2H2', 'FLH2O_17']
```

```python
g = sns.PairGrid(spitzer, dropna=True, vars=variables, diag_sharey=False, corner=True)
g.map_lower(sns.scatterplot)
g.map_lower(is_log, log_vars=plot_log)
```
