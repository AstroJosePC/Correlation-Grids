```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
sns.set_theme()
sns.set_style("ticks")
sns.set_context("paper")
```

```python
data_path = '../../Data/Spitzer_ALMA_sample.csv'
spitzer = pd.read_csv(data_path, sep=',', skipinitialspace=True, na_values=['#NAME?'])
plot_log = ['MSTAR', 'LSTAR', 'DIST', 'FLC2H2', 'FLH2O_17']
```

```python
x_variables = 'MSTAR', 'LSTAR', 'LOGLACC', 'LOGMACC', 'TEFF', 'N1331', 'DIST', \
#                'INCL', 'RSUBL', 'RSNOW_ACC', 'LOGRDUST68', 'R68_REF', 'LOGRDUST95'
y_variables = 'FLH2O_17', 'FLC2H2', 'F09MM', 'F13MM', 'FLH2O_17', 'FLH2O_33', \
#                'FLHCN', 'FLC2H2', 'FLCO2', 'FLH2_17'
```

```python
def is_log(x, y, log_vars=None, **kwargs):
    # Map log scales
    # pass
    # Get current axis
    ax = plt.gca()

    if x.name in log_vars:
        ax.set_xscale('log')
    if y.name in log_vars:
        ax.set_yscale('log')
    ax.autoscale()

g = sns.PairGrid(spitzer, x_vars=x_variables, y_vars=y_variables)
# g.map_diag(sns.histplot, color=".3")
g.map_offdiag(sns.scatterplot)
g.map_offdiag(is_log, log_vars=plot_log)
# g.add_legend()
```
